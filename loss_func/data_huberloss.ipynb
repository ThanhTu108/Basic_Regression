```python
def predict(x, w, b):
    return x*w +b

def compute_loss(y_hat, y , delta = 5):
    if abs(y_hat - y) < delta:
        loss = (y_hat - y)**2
    else:
        loss = delta*abs(y_hat - y) - 0.5*(delta**2)
    return loss
# square error
def compute_gradient_sr(y_hat, y, x):
    dw = 2*x*(y_hat - y)
    db = 2*(y_hat - y)
    return (dw, db)
# absolute
def compute_gradient_abs(y_hat, y, x, delta = 5):
    dw = delta*x*(y_hat - y) / abs(y_hat - y)
    db = delta*(y_hat - y) / abs(y_hat - y)
    return (dw, db)
def gradient(y_hat, y, x, delta = 5):
    if abs(y_hat - y) < delta:
        dw, db = compute_gradient_sr(y_hat, y, x)
    else:
        dw, db = compute_gradient_abs(y_hat, y, x, delta)
    return (dw, db)

def update_weights(w, b, lr, dw, db):
    w_new = w - lr*dw
    b_new = b - lr*db
    return (w_new, b_new)
```
